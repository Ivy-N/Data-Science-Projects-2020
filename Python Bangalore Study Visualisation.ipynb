```python
# necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
import seaborn as sns
```

```python
%matplotlib inline
```

```python
Gender = ['Females', 'Males']  
made_payments = [14, 26]  
no_payments = [86, 74]

plt.title("Percentage of Respondents who Made Digital Payments in the Past Year, grouped by gender (2017)");
plt.barh(Gender, made_payments, color="#333549", label="made_payments")  
# careful: notice "bottom" parameter became "left"
plt.barh(Gender, no_payments, left=made_payments, color="#3B4294", label="no_payments")

# we also need to switch the labels
plt.xlabel('Percentage of Respondents who Made Digital Payments in the Past Year, grouped by gender (2017)')  
plt.ylabel('Gender')

plt.legend()

plt.show() 
```

```python
data = {"Made digital payment in the past year":[14, 26],

        "Did not make digital payment in the past year":[86, 74]

        };

index     = ["Females", "Males"];

 

# Dictionary loaded into a DataFrame

dataFrame = pd.DataFrame(data=data, index=index);

 

# Draw a vertical bar chart

dataFrame.plot.barh(stacked=True,rot=15, title="Percentage of Respondents who Made Digital Payments in the Past Year, grouped by gender (2017)");

plt.show(block=True);
```

```python
# set width of bar
barWidth = 0.25
 
# set height of bar
classroom_training = [6, 6]
individualised_training = [11, 16]
 
# Set position of bar on X axis
r1 = np.arange(len(classroom_training))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, classroom_training, color='#333549', width=barWidth, edgecolor='white', label='before training')
plt.bar(r2, individualised_training, color='#3B4294', width=barWidth, edgecolor='white', label='after training')
 
# Add xticks on the middle of the group bars
plt.xlabel('Use of digital payments \n before and after different types of training received', fontweight='bold')
plt.ylabel('Percentage of participants \n using digital payment apps', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(classroom_training))], ['Classroom Training', 'Individualised Training'])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
